```python
#Datos equivocados
#titanic: calses: 1, 2, 3 <- valores permitidos. Valor equivocado: 0, 4, 5, etc.
#para modificar un registro: registro 1
#df.loc[1, "pclass"] = 3
#para modificar toda la columns
for i in df.index:
    if df.loc[i, "pclass"] > 3 or df.loc[i, "pclass"] < 1:
        if df.loc[i, "class"] == "Third":
            df.loc[i, "pclass"] = 3
        elif df.loc[i, "class"] == "Second":
            df.loc[i, "pclass"] = 2
        elif df.loc[i, "class"] == "First":
            df.loc[i, "pclass"] = 1
```

```python
df
```

```python
#Para borrar datos equivocados y eliminar el renglon completo
for i in df.index:
    if df.loc[i, "age"] < 5.0:
        print(i)
        df.drop(i, inplace = True)
```

```python
df
```

```python
#para eliminar renglones duplicados
print(df.duplicated())
```

```python
#para eliminar duplicados
df.drop_duplicates(inplace = True)
```
