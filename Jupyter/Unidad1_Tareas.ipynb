Distribuccion

Utilizando el método norm.rvs() del módulo scipy.stats se puede generar una distribución normal con variables aleatorias.
* loc = corresponde a la media de la distribución.
* scale = corresponde a la desviación estándar 
* size = al número de variantes aleatorias

```python
# import matplotlib.pyplot
import matplotlib.pyplot as plt
# import seaborn
import seaborn as sns
# import normal distribution
from scipy.stats import norm
```

```python
data_normal = norm.rvs(size=10000,loc=0,scale=1)
dis = sns.distplot(data_normal)
plt.show()
```

Estimación

Parámetros
* conjunto de datos: arreglo
      Puntos de datos para estimar. 

* bw_method: str (opcional)
      El método utilizado para calcular el ancho de banda del estimador. 

* weights: arreglo (opcional)
    pesos de los puntos de datos. Debe tener la misma forma que el conjunto de datos. 
    Por defecto, se asume que las muestras están igualmente ponderadas.

```python
# imports
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

# asignacion de valores random
m1 = np.random.normal(size=1000)
m2 = np.random.normal(scale=0.5, size=1000)

# operacion sobre las variables con los valores random
m1 = m1+m2 
m2 = m1-m2

# sacar el minimo y el maximo de 'x' y 'y'
xmin = m1.min()
xmax = m1.max()
ymin = m2.min()
ymax = m2.max()


# Creacion de una figura y una serie de subplots
fig, ax = plt.subplots()

        # valores - color - tamaño
ax.plot(m1, m2, 'b.', markersize=2)

# limites de 'x' y 'y'
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])

# mostrar grafica
plt.show()
```

Clustering

* vq


    Parámetros
    * obs: - arreglo
        Cada fila del array 'M' x 'N' es una observación. 

    * code_book: - ndarray
        code_book se genera normalmente utilizando el algoritmo k-means. 

    * check_finite: - bool (opcional)
        Si se comprueba que las matrices de entrada contienen sólo números finitos. 
        
    Regresa
    * code: - arreglo
        Una matriz de longitud M que contiene el índice del libro de códigos para cada observación.

    * dist: - arreglo
        La distensión (distancia) entre la observación y su código más cercano.


```python
# imports
from scipy.cluster.vq import kmeans, vq
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# variable con un diccionario con listas
d = {'x_coordinates' : [3.1, 2.2, 1.3, 1.2], 'y_coordinates' : [4.8, 9.6, 10.5, 11.3]}

# creacion de un dataframe con la variable como parametro
df = pd.DataFrame(data=d)

centroids,_ = kmeans(df,3)
df['cluster'],_ = vq(df,centroids)

sns.scatterplot(data=df)

# mostrar grafica
plt.show()
```
